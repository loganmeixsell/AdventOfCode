```python
with open(r'Input\Day2Input.txt') as f:
    lines = f.readlines()
lines
```

```python
mylist = []
for x in lines:
    mylist.append(x.split('\n')[0])
```

```python
# Part one
dictscore = {'X':1, 'Y':2, 'Z':3}
dictletter = {'X':'A', 'Y':'B', 'Z':'C'}
totalscore = 0
for x in mylist:
    handscore = dictscore[x[2]]
    myhand = dictletter[x[2]]
    print(x[0], x[2], handscore, myhand)
    if x[0] == myhand:
        print(f"Tie with a score of {handscore + 3}")
        totalscore += (handscore + 3)
    if x[0] == 'C' and myhand == 'A':
        # print("Win")
        print(f"Tie with a score of {handscore + 6}")
        totalscore += (handscore + 6)
    if x[0] == 'C' and myhand == 'B':
        # print("Loss")    
        print(f"Tie with a score of {handscore + 0}")
        totalscore += (handscore + 0)
    if x[0] == 'B' and myhand == 'A':
        # print("Loss")
        print(f"Tie with a score of {handscore + 0}")
        totalscore += (handscore + 0)
    if x[0] == 'B' and myhand == 'C':
        # print("Win")  
        print(f"Tie with a score of {handscore + 6}")
        totalscore += (handscore + 6) 
    if x[0] == 'A' and myhand == 'B':
        # print("Win")
        print(f"Tie with a score of {handscore + 6}")
        totalscore += (handscore + 6)
    if x[0] == 'A' and myhand == 'C':
        # print("Loss")   
        print(f"Tie with a score of {handscore + 0}")
        totalscore += (handscore + 0)
# A - Rock / B - Paper / C - Scissors
print(totalscore)
```

```python
# Part two
losepath = {'A':'C', 'B':'A', 'C':'B'}
winpath = {'A':'B', 'B':'C', 'C':'A'}
drawpath = {'A':'A', 'B':'B', 'C':'C'}
dictscore = {'X':1, 'Y':2, 'Z':3, 'A':1, 'B':2, 'C':3}
dictletter = {'X':'A', 'Y':'B', 'Z':'C'}
totalscore = 0
for x in mylist:
    # print(totalscore)
    handscore = dictscore[x[2]]
    # myhand = dictletter[x[2]]
    # print(x[0], x[2])
    if x[2] == 'X':
        # print(f"Lose {x[0]} against {losepath[x[0]]} ")
        letter = losepath[x[0]] 
        totalscore += (dictscore[letter] + 0)
    elif x[2] == 'Y':
        # print(f"Draw {x[0]} against {drawpath[x[0]]} ")
        letter = drawpath[x[0]] 
        totalscore += (dictscore[letter] + 3)
        # print("Draw")
    elif x[2] == 'Z':
        # print(f"Win {x[0]} against {winpath[x[0]]} ")
        letter = winpath[x[0]] 
        totalscore += (dictscore[letter] + 6)
        # print("Win")
    # handscore = dictscore[x[2]]
    # outcome = dictletter[x[2]]
print (totalscore)
```
